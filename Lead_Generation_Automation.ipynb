```python
import requests
import pandas as pd

# FETCHING 30 REAL USERS FROM API

url = "https://randomuser.me/api/?results=30"

response = requests.get(url)

data = response.json()

users = data["results"]

# EXTRACTING LEAD INFORMATION

lead_data = []

for user in users:

    # Full Name
    first_name = user["name"]["first"]
    last_name = user["name"]["last"]

    full_name = first_name + " " + last_name

    # Email
    email = user["email"]

    # Website (generated from username)
    username = user["login"]["username"]

    website = f"https://www.{username}.com"

    # Location
    city = user["location"]["city"]

    country = user["location"]["country"]

    location = city + ", " + country

    # Add to list
    lead = {
        "Name": full_name,
        "Email": email,
        "Website": website,
        "Location": location
    }

    lead_data.append(lead)

# CREATE DATAFRAME

df = pd.DataFrame(lead_data)

# DATA CLEANING

# Remove duplicate emails
df.drop_duplicates(subset=["Email"], inplace=True)

# Handle missing values
df.fillna("Not Available", inplace=True)

# SAVE TO EXCEL

output_file = "lead_generation_data.xlsx"

df.to_excel(output_file, index=False)

# OUTPUT

print("Lead Generation Completed Successfully!")
print(f"Total Leads Collected: {len(df)}")

display(df)
```
