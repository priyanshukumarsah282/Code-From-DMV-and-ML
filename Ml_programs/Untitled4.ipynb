```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = pd.read_csv("/content/single_regression.csv")
X = data[['YearsExperience']]
y = data['Salary']

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

plt.scatter(X, y, color='blue', label='Data points')

plt.plot(X, y_pred, color='red', label='Regression line')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load dataset
data = pd.read_csv("/content/mlr_housing.csv")

# Independent variables (multiple)
X = data[['area', 'bedrooms', 'bathrooms', 'stories']]

# Dependent variable
y = data['price']

# Train model
model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

plt.scatter(y, y_pred, color='blue', label="Data points")

# Perfect prediction line
min_val = min(min(y), min(y_pred))
max_val = max(max(y), max(y_pred))

plt.plot([min_val, max_val], [min_val, max_val], color='red', linewidth=2, label="Perfect Prediction")

plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted (Multiple Linear Regression)")
plt.legend()

plt.show()
```
