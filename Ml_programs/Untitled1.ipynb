```python
# ==============================
# VOICE DATASET EMOTION SORTER
# ==============================

import os
import shutil
from google.colab import drive

# 1️⃣ Mount Google Drive
drive.mount('/content/drive')

# 2️⃣ Dataset Path
BASE_PATH = "/content/drive/MyDrive/VoiceDataset"

# 3️⃣ Output Folder
OUTPUT_PATH = "/content/drive/MyDrive/VoiceDataset_Emotion_Sorted"
os.makedirs(OUTPUT_PATH, exist_ok=True)

# 4️⃣ Emotion Mapping

# RAVDESS emotion codes
ravdess_emotions = {
    "01": "neutral",
    "02": "calm",
    "03": "happy",
    "04": "sad",
    "05": "angry",
    "06": "fearful",
    "07": "disgust",
    "08": "surprised"
}

# CREMA-D emotion codes
crema_emotions = {
    "SAD": "sad",
    "ANG": "angry",
    "DIS": "disgust",
    "FEA": "fearful",
    "HAP": "happy",
    "NEU": "neutral"
}

# SAVEE emotion codes
savee_emotions = {
    "a": "angry",
    "d": "disgust",
    "f": "fearful",
    "h": "happy",
    "n": "neutral",
    "sa": "sad",
    "su": "surprised"
}

# 5️⃣ Function to copy files
def copy_file(src, emotion):
    emotion_folder = os.path.join(OUTPUT_PATH, emotion)
    os.makedirs(emotion_folder, exist_ok=True)
    shutil.copy(src, emotion_folder)

# 6️⃣ Process Files
for root, dirs, files in os.walk(BASE_PATH):
    for file in files:
        if file.endswith(".wav"):
            file_path = os.path.join(root, file)

            try:
                # -------- RAVDESS --------
                if "RAVDESS" in root:
                    parts = file.split("-")
                    emotion_code = parts[2]
                    emotion = ravdess_emotions.get(emotion_code)

                # -------- CREMA --------
                elif "Crema" in root or "CREMA" in root:
                    parts = file.split("_")
                    emotion_code = parts[2]
                    emotion = crema_emotions.get(emotion_code)

                # -------- SAVEE --------
                elif "Savee" in root or "SAVEE" in root:
                    name = file.split(".")[0]
                    if name[:2] in savee_emotions:
                        emotion = savee_emotions.get(name[:2])
                    else:
                        emotion = savee_emotions.get(name[0])

                # -------- TESS --------
                elif "Tess" in root or "TESS" in root:
                    emotion = root.split("_")[-1].lower()

                else:
                    emotion = None

                if emotion:
                    copy_file(file_path, emotion)

            except Exception as e:
                print("Error processing:", file_path)

print("✅ Dataset successfully separated by emotion!")
```
